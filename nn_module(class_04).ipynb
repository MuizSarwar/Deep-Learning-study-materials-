```python
import torch
import torch.nn as nn
```

```python
#dataset
features = torch.rand(10,5)   #dataset will have 5 features & 10 values in each.
features
```

#Version - 01 ( Single Layer )

```python
#----> A NN module with n inputes and 1 output <-----
class ModelVersion01(nn.Module):
  def __init__(self,feature_num):
    super().__init__()
    self.linear = nn.Linear(feature_num,1)
    self.sigmoid = nn.Sigmoid()


  def forward(self,features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out
```

```python
#create model
model_v1 = ModelVersion01(features.shape[1])

#forward pass
model_v1(features)
```

```python
#weights and bias of the model:-
model_weights = model_v1.linear.weight
model_bias = model_v1.linear.bias

print("Model weights:------------")
display(model_weights)
print("Model bias:---------------")
display(model_bias)
```

```python
from torchinfo import summary
summary(model_v1,input_size=features.shape)
```

#Version - 02 ( Multiple Layer )


```python
#-------------> layer:01 ----> input:feature_num  & output:3
#-------------> layer:02 ----> input:3 & output:1

class ModelVersion02(nn.Module):
  def __init__(self,feature_num):
    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(feature_num,3), #index=0
        nn.ReLU(),  #index=1
        nn.Linear(3,1), #index=2
        nn.Sigmoid()  #index = 3
    )


  def forward(self,features):
    out = self.network(features)
    return out
```

```python
from torch.return_types import mode
#create model
model_v2 = ModelVersion02(features.shape[1])

#forward pass
model_v2(features)
```

```python
#Weights and bias
layer01_weights = model_v2.network[0].weight
layer01_bias = model_v2.network[0].bias

layer02_weights = model_v2.network[2].weight
layer02_bias = model_v2.network[2].bias
```

```python
summary(model_v2,input_size=features.shape)
```
