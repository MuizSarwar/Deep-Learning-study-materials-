```python
import torch
```

#Gradient Calculation ( **Manual Process** )

```python
x = torch.tensor(6.7) #features
y = torch.tensor(0.0) #prediction

w = torch.tensor(1.0) #weight
b = torch.tensor(0.0) #bias
```

```python
def binary_cross_entropy_loss(prediction,target):
  epsilon = 1e-8
  prediction = torch.clamp(prediction,epsilon,1-epsilon)
  return -(y * torch.log(prediction) + (1-y) * torch.log(1-prediction))
```

```python
#Forward Pass
z = w*x + b
y_pred = torch.sigmoid(z)
L = binary_cross_entropy_loss(y_pred,y)
```

```python
#Derivatives
dL_dy_pred = (y_pred-y)/(y_pred*(1-y_pred))
dy_pred_dz = y_pred*(1-y_pred)
dz_dw = x
dz_db = 1
```

```python
#Gradient
dL_dw = dL_dy_pred * dy_pred_dz * dz_dw
dL_db = dL_dy_pred * dy_pred_dz * dz_db


display(dL_dw)
display(dL_db)
```

#Gradient Calculation ( **Using Autograd** )

```python
x = torch.tensor(6.7) #features
y = torch.tensor(0.0) #prediction

w = torch.tensor(1.0,requires_grad=True) #weight
b = torch.tensor(0.0,requires_grad=True) #bias
#----> We are calculating dL/dw  and dL/db , so we need to track w and b
```

```python
def binary_cross_entropy_loss(prediction,target):
  epsilon = 1e-8
  prediction = torch.clamp(prediction,epsilon,1-epsilon)
  return -(y * torch.log(prediction) + (1-y) * torch.log(1-prediction))
```

```python
#Forward Pass
z = w*x + b
y_pred = torch.sigmoid(z)
L = binary_cross_entropy_loss(y_pred,y)
```

```python
#Watch the track:
display(z)
display(y_pred)
display(L)
```

```python
#Derivative (using autograd)
L.backward()
```

```python
display(w.grad)  #----> Result of dL/dw  (Gradient for weight )
display(b.grad)  #----> Result of dL/db  (Gradient for bias )
```
