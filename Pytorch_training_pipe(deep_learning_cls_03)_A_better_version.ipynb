```python
#Libraries
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
#dataset load
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
#drop columns
df.drop(columns=['id','Unnamed: 32'],inplace=True)
df.head()
```

```python
#Target & Features
X = df.drop(['diagnosis'],axis=1) #Features
y = df['diagnosis'] #Target

#Train test split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
#Feature Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Label Encoding on Target
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
#Numpy array to Pytorch tensor :

X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
#data type conversion
X_train_tensor = torch.tensor(X_train_tensor,dtype=torch.float32)
X_test_tensor = torch.tensor(X_test_tensor,dtype=torch.float32)

y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
#data type conversion
y_train_tensor = torch.tensor(y_train_tensor,dtype=torch.float32)
y_test_tensor = torch.tensor(y_test_tensor,dtype=torch.float32)
```

```python
import torch.nn as nn
#Define the model:
class SimpleNN(nn.Module):
  def __init__(self,features_num):
    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(features_num,1),
        nn.Sigmoid()
    )

  def forward(self,features):
    out = self.network(features)
    return out
```

```python
#parameters
learning_rate = 0.1
epochs = 25
```

```python
loss_function = nn.BCELoss()
```

```python
#Training Pipeline
model = SimpleNN(X_train_tensor.shape[1])

#optimizer
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)

for epoch in range(epochs):
  #forward pass
  y_pred = model.forward(X_train_tensor)

  #loss
  loss = loss_function(y_pred,y_train_tensor.reshape(-1,1))

  #zero gradient
  optimizer.zero_grad()

  #backward
  loss.backward()

  #parameters update
  optimizer.step()


  # print loss in each epoch
  print(f'Epoch: {epoch + 1}, Loss: {loss}')
```

```python
#Evaluation
with torch.no_grad():
  y_pred = model.forward(X_test_tensor)
  y_pred = (y_pred > 0.75).float()
  accuracy = (y_pred == y_test_tensor).float().mean()  #Accuracy = (TP+TN)/(TP+TN+FP+FN)
  print(f'Accuracy: {accuracy.item()}')
```
